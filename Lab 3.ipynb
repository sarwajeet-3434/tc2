```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("iris.csv")
data.columns = data.columns.str.strip()
print("Dataset Summary:")
print(data.describe())
```

```python
#Calculate the mean of Sepal Length
sepal_length_mean = np.mean(data['sepal_length']) 
print(f"\nMean of Sepal Length: {sepal_length_mean:.2f}")

#Calculate the standard deviation of Sepal Width
sepal_width_std = np.std(data['sepal_width'])
print(f"Standard Deviation of Sepal Width: {sepal_width_std:.2f}")

#Calculate the correlation between Sepal Length and Sepal Width
correlation = np.corrcoef(data['sepal_length'], data['sepal_width'])[0, 1]
print(f"Correlation between Sepal Length and Sepal Width: {correlation:.2f}")
```

```python
plt.figure(figsize=(8, 6))  # Corrected the argument name

#Fixed syntax for histogram
plt.hist(data['sepal_length'], bins=20, color='skyblue', edgecolor="black")

#Fixed quotes and parentheses
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')

plt.grid(True)  # Ensures grid lines are displayed
plt.show()  # Displays the plot
```
