```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the Iris dataset
try:
    data = pd.read_csv("iris.csv")
    print("CSV file loaded successfully.")
except FileNotFoundError:
    print("Error: iris.csv file not found.")
    exit()

# Access the 'sepal_length' column
if "sepal_length" in data.columns:
    sepal_length_data = data["sepal_length"]
else:
    print("Error: 'sepal_length' column not found in the dataset.")
    exit()

# Sort the data
sorted_data = np.sort(sepal_length_data)

# Compute the CDF (Cumulative Distribution Function)
cdf = np.arange(1, len(sorted_data) + 1) / len(sorted_data)

# Plot the CDF
plt.figure(figsize=(8, 6))
plt.plot(sorted_data, cdf, marker='.', linestyle='none', color='red')
plt.title("Cumulative Distribution Function (CDF) for Sepal Length")
plt.xlabel('Sepal Length (cm)')
plt.ylabel('CDF')
plt.grid(True)
plt.show()

```

```python
variable = 'sepal_length'

#Ensure the variable name exists in the dataset
count, bins, ignored = plt.hist(data[variable], bins=20, density=True, color='skyblue', edgecolor="black")
bin_centers = 0.5 * (bins[:-1] + bins[1:])  # Calculate the centers of the bins

plt.figure(figsize=(8, 6))
plt.bar(bin_centers, count / sum(count), width=(bins[1] - bins[0]), color='lightgreen', edgecolor='black')

plt.title(f"PMF-like Visualization for {variable.title()}")
plt.xlabel(variable.title())
plt.ylabel('Probability')

plt.show()
```
