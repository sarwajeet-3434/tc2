```python
import pandas as pd
import numpy as np

def is_number(value):
    try:
        float(value)
        return True
    except ValueError:
        return False

def identify_column_types_from_dataframe(dataframe):
    categorical_columns, numerical_columns = set(), set()
    for column in dataframe.columns:
        is_numerical = True
        for value in dataframe[column]:
            if not is_number(value):
                is_numerical = False
                break
        if is_numerical:
            numerical_columns.add(column)
        else:
            categorical_columns.add(column)
    return categorical_columns, numerical_columns

data = pd.read_csv("iris.csv")
categorical, numerical = identify_column_types_from_dataframe(data)

print("Categorical Columns:", categorical)
print("Numerical Columns:", numerical)
```

```python
import pandas as pd

def compute_statistics(data):
    """Compute mean, variance, and standard deviation for a list of numbers."""
    if data.empty:
        print("Error: The data series is empty.")
        return None, None, None  # Check if the Series is empty

    n = len(data)
    mean = data.sum() / n
    variance = ((data - mean) ** 2).sum() / n  # Fixed: Use ** 2 for squaring the differences
    std_deviation = variance ** 0.5  # Fixed: Use ** 0.5 to compute square root

    return mean, variance, std_deviation

# Load the Iris dataset
try:
    data = pd.read_csv("iris.csv")
    print("CSV file loaded successfully.")
except FileNotFoundError:
    print("Error: iris.csv file not found.")
    exit()

# Display available columns for debugging
print("Available columns:", data.columns)

# Access the correct column (replace with the correct column name)
if "sepal_length" in data.columns:
    sepal_length_data = data["sepal_length"]
else:
    print("Error: 'sepal_length' column not found in the dataset.")
    exit()

# Compute statistics
mean, variance, std_dev = compute_statistics(sepal_length_data)

# Display results
if mean is not None:
    print(f"Mean: {mean:.2f}")
    print(f"Variance: {variance:.2f}")
    print(f"Standard Deviation: {std_dev:.2f}")
```
